```python
df["TenureGroup"] = pd.cut(df["tenure"],bins=[0,12,24,48,72],labels=["New Customers","Early Users","Mid-Term Customers","Loyal Customers"])
```

```python
df.groupby(["TenureGroup","Churn"]).size()
```

```python
#Customers in the first 12 months show the highest churn risk, indicating potential onboarding or early experience issues.
```

```python
tenure_churn = df.groupby(["TenureGroup","Churn"]).size().unstack()
```

```python
tenure_churn
```

```python
tenure_churn["ChurnRate"] = round((tenure_churn["Yes"] / (tenure_churn["Yes"] + tenure_churn["No"])) * 100, 2)
tenure_churn
```
